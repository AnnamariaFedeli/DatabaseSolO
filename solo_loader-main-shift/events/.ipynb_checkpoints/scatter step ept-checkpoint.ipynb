# STEP EPT

```python
data = pd.concat([fit_run, make_fit], axis=1)
data.query('fit_type == "best" and type_of_fit == "step_ept" and min_e_mev<0.03295087252 and max_e_mev >0.07803193193' , inplace = True)

one = data.where((data.averaging == 1), np.nan)
two = data.where((data.averaging == 2), np.nan)
five= data.where((data.averaging == 5), np.nan)

df_zero_err_one = one.where((data['break_point_err_mev'] == 0), np.nan)
df_zero_err_two = two.where((data['break_point_err_mev'] == 0), np.nan)
df_zero_err_five = five.where((data['break_point_err_mev'] == 0), np.nan)

data.query('break_point_err_mev != 0', inplace =True)
one.query('break_point_err_mev != 0', inplace =True)
two.query('break_point_err_mev != 0', inplace =True)
five.query('break_point_err_mev != 0', inplace =True)

n1 = str(one['gamma1'].count())
n2 = str(two['gamma1'].count())
n5 = str(five['gamma1'].count())

zero1 =  str(df_zero_err_one['gamma1'].count())
zero2 =  str(df_zero_err_two['gamma1'].count())
zero5 =  str(df_zero_err_five['gamma1'].count())

```

```python
f, ax = plt.subplots(3,1, figsize=(10,30)) 

ax[0].errorbar(x=one['break_point_mev'], y=one['gamma1'], yerr=one['gamma1_err'], xerr=one['break_point_err_mev'], color='red', fmt='o', markersize=5, ecolor='red', zorder=0, alpha=0.9)#, label='Flux average')
ax[0].errorbar(df_zero_err_one.break_point_mev, df_zero_err_one.gamma1,yerr=df_zero_err_one['gamma1_err'], xerr=df_zero_err_one['break_point_err_mev'], fmt='o', markersize=5, c='black', label='zero err', alpha=1)
ax[0].set_xlabel('Break energy [MeV]', size=12)
ax[0].set_ylabel('$\gamma_1$', size=12)
ax[0].set_title('One min av'+'\n'+'Events:  '+ n1+'  Zero err: '+ zero1)
ax[0].grid()

ax[1].errorbar(x=two['break_point_mev'], y=two['gamma1'], yerr=two['gamma1_err'], xerr=two['break_point_err_mev'], color='red', fmt='o', markersize=5, ecolor='red', zorder=0, alpha=0.9)#, label='Flux average')
ax[1].errorbar(df_zero_err_two.break_point_mev, df_zero_err_two.gamma1,yerr=df_zero_err_two['gamma1_err'], xerr=df_zero_err_two['break_point_err_mev'], fmt='o', markersize=5, c='black', label='zero err', alpha=1)
ax[1].set_xlabel('Break energy [MeV]', size=12)
ax[1].set_ylabel('$\gamma_1$', size=12)
ax[1].set_title('Two min av'+'\n'+'Events:  '+ n2+'  Zero err: '+ zero2)
ax[1].grid()

ax[2].errorbar(x=five['break_point_mev'], y=five['gamma1'], yerr=five['gamma1_err'], xerr=five['break_point_err_mev'], color='red', fmt='o', markersize=5, ecolor='red', zorder=0, alpha=0.9)#, label='Flux average')
ax[2].errorbar(df_zero_err_five.break_point_mev, df_zero_err_five.gamma1,yerr=df_zero_err_five['gamma1_err'], xerr=df_zero_err_five['break_point_err_mev'], fmt='o', markersize=5, c='black', label='zero err', alpha=1)
ax[2].set_xlabel('Break energy [MeV]', size=12)
ax[2].set_ylabel('$\gamma_1$', size=12)
ax[2].set_title('Five min av'+'\n'+'Events:  '+ n5+'  Zero err: '+ zero5)
ax[2].grid()

plt.tick_params(axis='x', which='minor', labelsize=10)
#ax.xaxis.set_minor_formatter(FormatStrFormatter("%.2f"))
    #plt.tick_params(axis='y', which='minor')
    #ax.yaxis.set_minor_formatter(FormatStrFormatter("%.0f"))
plt.legend(prop={'size': 10})
plt.xticks(size=10)
plt.yticks(size=10)


#ax.set_ylim(-20,0)
#plt.title('Events:  '+ number+'  Zero err: '+ zero)
plt.show()
```
